```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Ссылка на данные https://archive.ics.uci.edu/ml/datasets/iris

```python
iris = pd.read_csv('C:/Users/user/Downloads/iris.csv', index_col = 'Id')
```

Посмотрим на данные

```python
iris.head(10)
```

Описательная статистика:

```python
iris.describe()
```

Посмотрим на связи переменных между собой, для этого построим тепловую карту:

```python
fig, ax = plt.subplots(figsize=(5,5))
ax = sns.heatmap(iris.corr(), annot=True, linewidths=.5, square=True, ax= ax)
i, k = ax.get_ylim()
ax.set_ylim(i+0.5, k-0.5)
```

Наибольшая зависимость наблюдается у признаков PetalLength и PetalWidth (0.96) - прямая зависимость, то есть чем больше длина внутренней доли околоцветника ириса (PetalLength), тем больше ширина внутренней доли околоцветника (PetalWidth).

Диаграммы рассеяния для визуализации корреляции между признаками:

```python
sns.pairplot(iris, hue = 'Species', height = 3)
plt.show()
```
